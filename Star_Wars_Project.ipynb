````python
yes_no = { "Yes" : True,
         "No" : False}
````

````python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
````

````python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
````

````python
star_wars.head(10)
````

Convert each column above so that it only contains the values ```True``` and ```False```.

````python
import numpy as np
star_wars_map = {"Star Wars: Episode I  The Phantom Menace": True,
                 "Star Wars: Episode II Attack of the Clones":True,
                 "Star Wars: Episode III Revenge of the Sith":True,
                 "Star Wars: Episode IV A New Hope":True,
                 "Star Wars: Episode V The Empire Strikes Back":True,
                 "Star Wars: Episode VI Return of the Jedi":True,
                 np.NaN: False}
````

````python
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(star_wars_map)
````

Rename each of the columns so the names are more intuitive.

````python
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
                                     "Unnamed: 4":"seen_2","Unnamed: 5":"seen_3","Unnamed: 6":"seen_4","Unnamed: 7":"seen_5","Unnamed: 8":"seen_6"})
````

````python
star_wars.head()
````

Convert each of the columns above to a ```float``` type.

````python
star_wars[pd.notnull(star_wars["RespondentID"])]
````

````python
#Give each column a more descriptive name.


star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })


star_wars.head()
````

````python
star_wars[pd.notnull(star_wars["RespondentID"])]
````

````python
star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.loc[1:,star_wars.columns[9:15]] = star_wars.loc[1:,star_wars.columns[9:15]].astype(float)
````

````python
#Compute the mean of each of the ranking column
star_wars.loc[1:,star_wars.columns[9:15]].mean()
````

````python
#Make a bar chart for each ranking.
import matplotlib.pyplot as plt
%matplotlib inline
star_wars.loc[1:,star_wars.columns[9:15]].mean().plot.bar()
````

Hence the average ranking for movie 3 is the highest which denotes that it did not do well. Movie 5 has the lowest ranking which means it did the most impressive. Hence **Star Wars: Episode III Revenge of the Sith** did the most impressively.

````python
# We can sum the "seen" columns to determine how many people watched the movie.

star_wars.loc[1:,star_wars.columns[3:9]].sum()
````

````python
star_wars.loc[1:,star_wars.columns[3:9]].sum().plot.bar()
````

This bar plot shows that no one has watched movie 2,3 and 4. The highest watched movie is movie 5 which is the **Star Wars: Episode III Revenge of the Sith** which probably explains why it did really well on the big screen.

````python
# Splitting the dataframe in two groups based on the gender
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
````

````python
#Doing all the analysis done above again for the two groups
males.loc[1:,males.columns[9:15]].mean()
````

````python
males.loc[1:,males.columns[9:15]].mean().plot.bar()
````

Hence among males we get a similar result for the movie which does the best.

````python
females.loc[1:,females.columns[9:15]].mean()
````

````python
females.loc[1:,females.columns[9:15]].mean().plot.bar()
````

Similar results hold among women too for the best movie
